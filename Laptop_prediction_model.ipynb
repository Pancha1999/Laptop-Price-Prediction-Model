```python
def add_company(inpt):
    if inpt=='Samsung' or inpt=='Razer' or inpt=='Mediacom' or inpt=='Microsoft' or inpt=='Xiaomi' or inpt=='Vero' or inpt=='Chuwi' or inpt=='Google' or inpt=='Fujitsu' or inpt=='LG' or inpt=='Huawei':
        return 'Other'
    else:
        return inpt
df['Company']=df['Company'].apply(add_company)
```

```python
df['Touchscreen']=df['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
df['Ips']=df['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
```

```python
df.head(5)
```

```python
df['Cpu'].value_counts()
```

```python
len(df['Cpu'].value_counts())
```

```python
df['Cpu_name']=df['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
def set_processor(name):
    if name=='Intel Core i7' or name=='Intel Core i5' or name=='Intel Core i3':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'

df['Cpu_name']=df['Cpu_name'].apply(set_processor)
```

```python
df['Cpu_name'].value_counts()
```

```python
df['Gpu'].value_counts()
```

```python
df['Gpu_name']=df['Gpu'].apply(lambda x:' '.join(x.split()[0:1]))
```

```python
df['Gpu_name'].value_counts()
```

```python
df.shape
```

```python
df = df[df['Gpu_name'] != 'ARM']
```

```python
df.shape
```

```python
df.head(5)
```

```python
df['OpSys'].value_counts()
```

```python
def os_type(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return 'Linux'
    else:
        return 'Other'
df['OpSys']=df['OpSys'].apply(os_type)
```

```python
df['OpSys'].value_counts()
```

```python
df.head(5)
```

```python
df.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'],inplace=True)
```

```python
df.head(5)
```

```python
df=pd.get_dummies(df)
```

```python
df=df.astype('int32')
```

```python
df.head(5)
```

```python
df.shape
```

```python
X=df.drop(columns=['Price_euros'],axis=1)
y=df['Price_euros']
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape,X_test.shape
```

```python
def model_acc(model):
    model.fit(X_train,y_train)
    acc=model.score(X_test,y_test)
    print(str(model)+'-->'+str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso=Lasso()
model_acc(lasso)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
model_acc(knn)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_acc(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'n_estimators':[10,20,30],'criterion':['squared_error','absolute_error','poisson']}
```

```python
grid_obj=GridSearchCV(estimator=rf,param_grid=parameters)
grid_fit=grid_obj.fit(X_train,y_train)
best_model=grid_fit.best_estimator_
```

```python
best_model
```

```python
best_model.score(X_test,y_test)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model,file)
```
